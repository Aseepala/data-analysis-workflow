```python
# ---------------------------------------------------------------
# Step 2: Load the Dataset
# ---------------------------------------------------------------
# Load the synthetic M365 Copilot issues dataset.
# The dataset contains 40 rows of user-reported issues across
# Word, Teams, Outlook, Excel, PowerPoint and Business Chat,
# sourced from UserVoice and Support Tickets.
df = pd.read_csv("../test_data/issues.csv")
print(f"Shape: {df.shape}")  # Print number of rows and columns
print(f"Columns: {list(df.columns)}")
df.head()  # Preview the first 5 rows
```

```python
# ---------------------------------------------------------------
# Step 3: Issues by Product Area
# ---------------------------------------------------------------
# Count how many issues each M365 product area has and plot as a bar chart.
# This tells us which products are generating the most user complaints.
df["product_area"].value_counts().plot(
    kind="bar", figsize=(10, 5), title="Issues by Product Area", color="steelblue"
)
plt.ylabel("Number of Issues")
plt.xlabel("Product Area")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# ---------------------------------------------------------------
# Step 4: Issues by Source
# ---------------------------------------------------------------
# Break down issues by where they were reported:
# - UserVoice: voluntary user feedback submissions
# - Support Ticket: issues raised via official support channels
# A pie chart shows the proportion of each source.
df["source"].value_counts().plot(
    kind="pie", figsize=(6, 6), title="Issues by Source", autopct="%1.1f%%"
)
plt.ylabel("")  # Remove the default y-axis label for cleaner display
plt.show()
```

```python
# ---------------------------------------------------------------
# Step 5: Issues Over Time
# ---------------------------------------------------------------
# Convert the date_reported column to a proper datetime type,
# then group by date and count how many issues were reported each day.
# This reveals any spikes or trends in issue reporting over time.
df["date_reported"] = pd.to_datetime(df["date_reported"])
df.groupby("date_reported").size().plot(
    kind="line", figsize=(12, 4), title="Issues Reported Over Time",
    marker="o", color="steelblue"
)
plt.ylabel("Number of Issues")
plt.xlabel("Date")
plt.tight_layout()
plt.show()
```

```python
# ---------------------------------------------------------------
# Step 6: Product Area vs Source Heatmap
# ---------------------------------------------------------------
# Create a pivot table that cross-references product area (rows)
# with source (columns) and counts the number of issues at each intersection.
# The heatmap makes it easy to spot which product + source combinations
# are most problematic.
pivot = df.pivot_table(index="product_area", columns="source", aggfunc="size", fill_value=0)
sns.heatmap(pivot, annot=True, fmt="d", cmap="Blues")  # fmt="d" = show integers
plt.title("Issues by Product Area and Source")
plt.tight_layout()
plt.show()
```
