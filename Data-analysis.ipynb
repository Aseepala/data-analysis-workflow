```python
# Load Data
df = pd.read_csv("issues.csv")
print(f"Shape: {df.shape}")
df.head()
```

```python
# Issues by Product Area
df["product_area"].value_counts().plot(
    kind="bar", figsize=(10, 5), title="Issues by Product Area", color="steelblue"
)
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Issues by Source (UserVoice vs Support Ticket)
df["source"].value_counts().plot(
    kind="pie", figsize=(6, 6), title="Issues by Source", autopct="%1.1f%%"
)
plt.ylabel("")
plt.show()
```

```python
# Issues Over Time
df["date_reported"] = pd.to_datetime(df["date_reported"])
df.groupby("date_reported").size().plot(
    kind="line", figsize=(12, 4), title="Issues Reported Over Time", marker="o", color="steelblue"
)
plt.ylabel("Number of Issues")
plt.tight_layout()
plt.show()
```
